```python
#Changing pandas dataframe to numpy array
X = df.iloc[:,:35].values
y = df.iloc[:,35:36].values
print(X.shape)
print(y.shape)
```

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(k_neighbors=15)
X, y = smote.fit_resample(X, y)
y
```

```python
from keras.utils import np_utils
from sklearn.preprocessing import LabelEncoder
# encode class values as integers
#encoder = LabelEncoder()
#encoder.fit(y)
#encoded_Y = encoder.transform(y)
# convert integers to dummy variables (i.e. one hot encoded)
dummy_y = np_utils.to_categorical(y)
dummy_y
```

### 0 - Drop out: 1 0 0; 1 - Graduate: 0 1 0; 2 - Enrolled: 0 0 1

## Building Model

#### Train Test Split

```python
#Normalizing the data
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, dummy_y, test_size = 0.2, random_state = 87)
```

```python
print(X_train.shape)
print(y_train.shape)
```

#### Creating Model

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Dropout
from tensorflow.keras.callbacks import EarlyStopping, TensorBoard
```

```python
log_directory = 'logs\\fit'

board = TensorBoard(log_dir = log_directory, histogram_freq = 1,
    write_graph = True,
    write_images = True,
    update_freq = 'epoch',
    profile_batch = 2,
    embeddings_freq = 1)
```

```python
df.info()
```

```python
model = Sequential()

model.add(Dense(units=512, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(units=512, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(units=512, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(units=512, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(units=512, activation='relu'))
model.add(Dropout(0.2))

model.add(Dense(units=3, activation='softmax'))

# compiler
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(x=X_train,y=y_train,epochs=500,validation_data=(X_test,y_test),batch_size=50)
```
